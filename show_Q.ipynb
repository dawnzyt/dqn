```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

q1, q_fix1 = np.load('qs1.npy'), np.load('qs_fix1.npy')
q2, q_fix2 = np.load('qs2.npy'), np.load('qs_fix2.npy')
l1 = min(len(q1), len(q2))
q1, q2 = q1[:l1], q2[:l1]
l2 = min(len(q_fix1), len(q_fix2))
q_fix1, q_fix2 = q_fix1[:l2], q_fix2[:l2]

# fix state [0.2,0], action: 0
df = pd.DataFrame({'dqn': q1, 'ddqn': q2, "x": range(len(q1))})
df_fix = pd.DataFrame({'dqn': q_fix1, 'ddqn': q_fix2, "x": range(len(q_fix1))})
```

```python
plt.figure(figsize=(10, 5))
# plt.plot('x', 'dqn', data=df, marker='', color='blue', linewidth=2, label="DQN")
# plt.plot('x', 'ddqn', data=df, marker='', color='red', linewidth=2, label="DDQN")
plt.plot('x', 'dqn', data=df_fix, marker='', color='blue', linewidth=2, linestyle='dashed')
plt.plot('x', 'ddqn', data=df_fix, marker='', color='red', linewidth=2, linestyle='dashed')
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Q-value")
plt.title("Q-value, fix state [0.2,0], action: 0")
plt.savefig("q_value.png")
plt.show()
```
